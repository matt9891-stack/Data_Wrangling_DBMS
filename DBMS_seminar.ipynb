# Connection to Postgre - Create a DB - Create a Table and Fill with Values

```python
import psycopg2
```

```python
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT
```

```python
connection = psycopg2.connect( #connection to POSTGRE through psycopg2 
    host='localhost',
    user = 'postgres',
    password = 'password'
)
```

```python
connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT) #creation of a new DB

cur = connection.cursor()

db_name = 'seminar'

cur.execute(f'CREATE DATABASE {db_name}')

cur.close()
connection.close()

print(f'Database {db_name} successfully created')
```

```python
#Connection to the DB

connection2 = psycopg2.connect(
    host='localhost',
    user='postgres',
    password='password',
    dbname='seminar'
)
```

```python
new_df1.columns
```

```python
# creation of the table that will host our values

cur2 = connection2.cursor()

# 3️⃣ Creazione tabella
query = """
CREATE TABLE child_labour (
Region VARCHAR(254),
Country VARCHAR(254),
Total_pct VARCHAR(254),
y VARCHAR(254),
Sex_pct_Male VARCHAR(254),
Sex_pct_Female VARCHAR(254),
Place_of_residence_Urban VARCHAR(254),
Placde_of_residence_Rural VARCHAR(254),
Household_wealth_quintile_Poorest VARCHAR(254),
Household_wealth_quintile_Second VARCHAR(254),
Household_wealth_quintile_Middle VARCHAR(254),
Household_wealth_quintile_Fourth VARCHAR(254),
Household_wealth_quintile_Richest VARCHAR(254),
Reference_Year VARCHAR(254),
Data_Source VARCHAR(254)
);
"""

cur2.execute(query)
connection2.commit()

cur2.close()
connection2.close()

print("Table successfully created")
```
